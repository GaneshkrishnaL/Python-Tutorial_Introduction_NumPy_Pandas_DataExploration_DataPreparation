## DATA EXPLORATION

Data exploration refers to the preliminary investigation of data in order to better understand its specific characteristics.

lets import a dataset and print irs head

```python
import pandas as pd

data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
data.columns=['sepal length','sepal width','petal lenth','petal width','class']
data.head()
```

lets see its index, columns, size, and shape, and dtypes

```python
data.index
```

```python
data.columns
```

```python
data.size
```

```python
data.shape
```

```python
data.dtypes
```

Summary statistics on individual columns and  overall dataset

- Summary statistics are quantities, such as the mean and standard deviation, that capture various characteristics of a potentially large set of values with a single number or a small set of numbers.

```python
data['sepal length'].describe()
```

-  to display the summary for all the attributes simultaneously in a table using the **describe()**  function. If an attribute is quantitative, it will display its mean, standard deviation and various quantiles (including minimum, median, and maximum) values. If an attribute is qualitative, it will display its number of unique values and the top (most frequent) values.

```python
data.describe()
```

lets see unique values per column and count of unique values

```python
data['class'].unique()
```

```python
data['sepal length'].nunique()
```

Getting the count of class using value_counts()

```python
data['class'].value_counts()
```

For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.

Covariance

```python
data.cov()
```

Correlation

```python
data.corr()
```

```python
#checking null values
data.isnull().sum()
```

```python
# get records where 'class' is 'Iris-setosa' and (&) 'sepal length' is 5.0

data[(data['class']=='Iris-setosa') & (data['sepal length']==5.0)]
```

DATA VISUALIZATION

- In this section, we'll see how to visualize the Iris data created above.

1. First, we will display the histogram for the sepal length attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

```python
%matplotlib inline

data['sepal length'].hist(bins=8)
```

Boxplot

- A boxplot can also be used to show the distribution of values for each attribute.

```python
data.boxplot()
```

Scatter Plot

- For each pair of attributes, we can use a scatter plot to visualize their joint distribution.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.pairplot(data)
```

```python

#we can  use Seaborn's pairplot and hue to distinguish classes with different colors


sns.pairplot(data,hue='class')
```

Parallel coordinates


- Parallel coordinates can be used to display all the data points simultaneously. An object is represented as a line instead of as a point. In the example below, lets see the distribution of values for each class to be identified in a separate color.

```python
from pandas.plotting import parallel_coordinates

parallel_coordinates(data,'class')
```
