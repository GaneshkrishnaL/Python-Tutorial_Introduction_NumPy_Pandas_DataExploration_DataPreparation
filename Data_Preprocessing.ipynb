##Handling missing values

It is not unusual for an object to be missing one or more attribute values. In some cases, the information was not collected; while in other cases, some attributes are inapplicable to the data instances. This section presents examples on the different approaches for handling missing values.

```python
cancer.dtypes
```

```python
cancer['Bare_nuclei'].unique()
```

```python
import numpy as np
cancer=cancer.replace('?',np.NaN)
```

```python
cancer['Bare_nuclei'].unique()
```

```python
cancer.isnull().sum()
```

###Different ways to handle Misisng data

1.Drooping the records that contain missing values

```python
cancer1=cancer.copy()
```

```python
cancer1.dropna(inplace=True)
```

```python
print(cancer.shape)
print(cancer1.shape)
```

2. Using Arbitrary value Imputation

```python
# replacing null values based on the user preferred arbitrary value
cancer2=cancer.copy()
cancer2.fillna(2,inplace=True)
```

```python
cancer[cancer['Bare_nuclei'].isnull()]
```

```python
cancer2.iloc[40]
```

3. Using Mean/ Median/ Mode Imputation

```python
cancer3=cancer.copy()
cancer3['Bare_nuclei'].fillna(int(cancer3['Bare_nuclei'].median()),inplace=True)
```

```python
cancer3.iloc[23]
```

```python
cancer3['Bare_nuclei'] = cancer3['Bare_nuclei'].astype(int)
```

```python
cancer3.dtypes
```

##  OUTLIERS
-  Outliers are data instances with characteristics that are considerably different from the rest of the dataset.
- most common techniques for finding outliers are Z scores and Inter Quartile Range.



Handling outliers using Z Score

```python
#box plot to identify outliers
```

```python
#Drop the class attribute
cancer3.drop(columns='class',inplace=True)
```

```python
cancer3.boxplot(figsize=(12,3))
```

```python
cancer3.mean()
```

```python
cancer3.std()
```

```python
Z=(cancer3-cancer3.mean())/cancer3.std()
```

```python
Z[30:40]
```

```python
Z2 = Z[(Z > -3) & (Z <= 3)].dropna()

# Display the number of rows before and after discarding outliers
print('Number of rows before discarding outliers = %d' % Z.shape[0])
print('Number of rows after discarding outliers = %d' % Z2.shape[0])
```

## Duplicate Data
- Some datasets, especially those obtained by merging multiple data sources, may contain duplicates or near duplicate instances.
- The term deduplication is often used to refer to the process of dealing with duplicate data issues.

Handling duplicate data

```python
dups = cancer.duplicated().sum()
```

```python
print('total duplicates =',dups)
```

```python
#dropping duplicates
print('total data before dropping dups = ',cancer.shape[0],'\n')
dropdups=cancer.drop_duplicates()
print('total data after dropping dups = ',dropdups.shape[0])
```

#2.Aggregation

Data aggregation is a preprocessing task where the values of two or more objects are combined into a single object. The motivation for aggregation includes:
-  reducing the size of data to be processed.

##Sampling

Sampling is an approach commonly used to facilitate data reduction for exploratory data analysis and scaling up algorithms to big data applications and quantifying uncertainties due to varying data distributions.

There are various methods available for data sampling, such as

- sampling without replacement, where each selected instance is removed from the dataset
- sampling with replacement, where each selected instance is not removed, thus allowing it to be selected more than once in the sample.

DISCRETIZATION

Discretization is a data preprocessing step that is often used to transform a continuous-valued attribute to a categorical attribute. The example below illustrates two simple but widely-used unsupervised discretization methods (equal width and equal depth) applied to the 'Clump Thickness' attribute of the breast cancer dataset.

```python
cancer['Clump_thickness'].hist(bins=10)
```

1 equal width : cut()
2.equal depth : qcut()

```python
bins=pd.cut(cancer['Clump_thickness'],4)
```

```python
bins.value_counts(sort=False)
```

```python
qbins=pd.qcut(cancer['Clump_thickness'],4)
qbins.value_counts(sort=False)
```

PRINCIPAL COMPONENT ANALYSIS

Principal component analysis (PCA) is a classical method for reducing the number of attributes in the data by projecting the data from its original high-dimensional space into a lower-dimensional space

Lets see how we can reduce 9 attribute data to 2 attributes and plot the data against class label.

```python
cancer
```

```python
classs=cancer['class']
```

```python
classs
```

```python
cancer.drop('class',axis=1, inplace=True)
```

```python
cancer
```

```python
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
pca.fit(cancer)

pc=pca.transform(cancer)
```

```python
cancer.shape
```

```python
pc.shape
```

```python
pc
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))

plt.scatter(pc[:,0],pc[:,1],c=classs)

plt.xlabel('pc1')
plt.ylabel('pc2')
```
